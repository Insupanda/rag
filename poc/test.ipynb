```python
import os
from src.models.collection_loader import CollectionLoader
from src.models.embeddings import UpstageEmbedding
from dotenv import load_dotenv
from src.models.search import search
from src.models.generate_answer import generate_answer
from src.util.utils import find_matching_collections

vector_path = "insu_data"
collections = []
query = "외상후 스트레스 장애(PTSD)를 보장하는 보험은?"
loader = CollectionLoader(vector_path, UpstageEmbedding)

available_collections = [
        d
        for d in os.listdir(loader.base_path)
        if os.path.isdir(os.path.join(loader.base_path, d))
    ]
print(f"사용 가능한 컬렉션: {available_collections}")

# 요청된 컬렉션이 있거나 쿼리 기반으로 컬렉션을 찾습니다
use_collections = (
    collections
    if collections
    else find_matching_collections(query, available_collections)
)
print(f"사용할 컬렉션: {use_collections}")

# 찾은 컬렉션 로드
for collection_name in use_collections:
    total_collections = loader.load_collection(collection_name)

# 청크 탐색 (각 컬렉션당 top_k=2)
search_results = search(query, total_collections, use_collections, top_k=2)

# 답변 생성
answer = generate_answer(
    query, search_results, os.getenv("OPENAI_API_KEY")
)

print(answer)
```
